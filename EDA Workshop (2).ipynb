```python
Rural_df=pd.DataFrame()
SemiU_df=pd.DataFrame()
Urban_df=pd.DataFrame()
```

```python
import matplotlib.pyplot as plt
import numpy as np
plt.figure(figsize=(8,6))
x=loan_data['Dependents'].value_counts()
print(x)
plt.xlabel('Number of Dependents')
plt.ylabel('Count')
plt.title('Distribution of Dependents')

plt.bar(x.index, x)
```

```python
x = loan_data['Dependents'].value_counts()
plt.figure(figsize=(8,6))
plt.pie(x, labels=x.index, autopct='%1.1f%%')
plt.title('Distribution of Dependents')
plt.show()
```

```python
counts = loan_data.groupby(['Education', 'Gender']).size().unstack()
counts.plot(kind='bar', color=['skyblue', 'pink'], figsize=(8, 6))

plt.title('Count of Education by Gender')
plt.xlabel('Education')
plt.ylabel('Count')
plt.xticks([0,1], ['nongrad', 'grad'])
plt.legend(['female', 'male'])
plt.show()
```

```python
counts = loan_data.groupby(['Credit_History', 'Loan_Status']).size().unstack()
counts.plot(kind='bar', color=['skyblue', 'pink'], figsize=(8, 6))

plt.title('Credit history based on marital status')
plt.xlabel('Marrital Status')
plt.ylabel('Count')
plt.xticks([0,1], ['nongrad', 'grad'])
plt.legend(['not approved', 'approved'])
plt.show()
```

```python
average_income = loan_data.groupby('Self_Employed')['Income'].mean()
average_income.plot(kind='bar', color ='skyblue', edgecolor='black')

plt.title('Average Income by Self-Employmen Status')
plt.xlabel('Self_Employed')
plt.ylabel('Average_Income')
plt.show()
```

```python
plt.hist(loan_data['Income'], bins=13, color = 'skyblue', edgecolor='black')

plt.title('Income Distribution')
plt.xlabel('Income')
plt.ylabel('Frequency')

plt.show()
```
