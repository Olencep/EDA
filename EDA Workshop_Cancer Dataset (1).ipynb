```python
m = df['Breast'].mode()[0]
df['Breast'] = df['Breast'].fillna(m)
```

```python
m = df['Tumor Size (cm)'].mean()
df['Tumor Size (cm)'] = df['Tumor Size (cm)'].fillna(m)
```

```python
m = df['Inv-Nodes'].mean()
df['Inv-Nodes'] = df['Inv-Nodes'].fillna(m)
```

```python
m = df['Metastasis'].mean()
df['Metastasis'] = df['Metastasis'].fillna(m)
```

```python
m = df['Breast Quadrant'].mode()[0]
df['Breast Quadrant'] = df['Breast Quadrant'].fillna(m)
```

```python
m = df['History'].mode()[0]
df['History'] = df['History'].fillna(m)
```

```python
df.isna().sum()
```

```python
df['Breast'].replace({'Right':1,'Left':0}, inplace=True)
#df['Breast Quadrant']=df['Breast Quadrant'].replace({'Upper inner':1,'Lower inner':0,'Upper outer':2,'Lower outer':-1})
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
relation = df.groupby('Age')['Tumor Size (cm)'].mean()
relation.plot(kind='line', color ='skyblue')

plt.title('Relationship between Tumor Size and Age')
plt.ylabel('Tumor Size (cm)')
plt.xlabel('Age')
plt.show()
```
